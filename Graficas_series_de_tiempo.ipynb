```python
# load time series dataset
import pandas as pd 
import numpy as np 
import seaborn as sns
```

```python
url='https://raw.githubusercontent.com/terranigmark/curso-analisis-exploratorio-datos-platzi/main/df_power.csv'
df_power_system= pd.read_csv(url,error_bad_lines=False)
```

```python
df_power_system.columns
```

```python
df_power_system.info()
```

```python
df_power_system['Date'] = pd.to_datetime(df_power_system['Date'])
```

```python
df_power_system = df_power_system.set_index('Date')
```

```python
df_power_system['Year'] = df_power_system.index.year
```

```python
df_power_system['Month'] = df_power_system.index.month

df_power_system['Day'] = df_power_system.index.day_name
```

```python
df_power_system.head()
```

```python
import matplotlib.pyplot as plt
```

```python
df_power_system['Consumption'].plot(linewidth=0.5)
```

```python
df_power_system.info()
```

```python
name= ['Consumption','Wind','Solar']
df_power_system[name].plot(marker= '.', figsize= (14,6), subplots= True, alpha= 0.5, linestyle= 'None')
```

```python
df_power_system.loc['2016','Consumption'].plot()
```

```python
df_power_system.loc['2016-12','Consumption'].plot(marker='o')
```

```python
fig, axess= plt.subplots(3,1,figsize=(8,7))
for name, ax in zip(['Consumption','Wind','Solar'],axess):
  sns.boxenplot(data= df_power_system, x='Month', y= name, ax= ax)
```

```python
fig, axess= plt.subplots(3,1,figsize=(8,7))
for name, ax in zip(['Consumption','Wind','Solar'],axess):
  sns.boxenplot(data= df_power_system, x='Year', y= name, ax= ax)
```
